```python

import numpy as np
import matplotlib.pyplot as plt

x = np.array([-40,0,80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([0,80,240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.figure(figsize=(10,10))

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.plot(x,y)

plt.grid()

plt.show()
```

shear transformation by 2 

```python
xhat = np.array([])
yhat = np.array([])

xhat= xhat.astype(int)
yhat= yhat.astype(int)

for i in range(0,x.size):
  xhat= np.append(xhat,(x[i]*1+y[i]*2))
  yhat= np.append(yhat,(y[i]*1))


plt.figure(figsize=(10,10))

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.plot(x,y,'r')
plt.plot(xhat,yhat)
plt.grid()

plt.show()
```

shear tranformation using matrix multiplication.

```python
M = np.array([x,y])
T = np.array([[1,2],[0,1]])

C = T.dot(M)

plt.figure(figsize=(10,10))

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.plot(x,y,'r')
plt.plot(C[0],C[1])
plt.grid()

plt.show()
```
